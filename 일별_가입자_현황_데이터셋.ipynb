#### (2) 회원 가입, 회원 탈퇴

```python
fig = plt.figure(figsize=(20,10)) ## 캔버스 생성
fig.set_facecolor('white')
ax1 = fig.add_subplot() ## axes 생성
 
color1 = 'b'
ax1.plot(data.연월일, data.회원가입, color=color1)
ax1.set_xlabel('x')
ax1.set_ylabel('회원가입', color=color1)
ax1.tick_params(axis='y', labelcolor=color1)
 
color2 = 'r'
ax2 = ax1.twinx()
ax2.plot(data.연월일, data.회원탈퇴, color=color2)
ax2.set_ylabel('회원탈퇴', color=color2)
ax2.tick_params(axis='y', labelcolor=color2)
 
plt.show()
```

* 로그 스케일 변환 → 분산을 일정하게 변환

p-value가 0.05이하이므로 대립가설 채택. 공적분이 있음으로 판단 → 일별 접속자수와 회원탈퇴자수는 공적분 관계에 있다. 

```python
(data_plot['접속자'] - (data_plot['회원가입'] + data_plot['회원탈퇴']) < 0).sum()
# 데이터 타당성 확인 : 잘못 입력된 값은 없는 것으로 파악
```

```python
(data_plot['접속자'] < data_plot['회원가입']).sum() 
# 데이터 타당성 확인 (모순이 있는지 확인작업)
# 접속을 하고 👉 바로 회원가입을 하는 경우 / 하지 않는 경우 
```

```python
data_plot['서비스이용접속'] = data_plot['접속자'] - (data_plot['회원가입'] + data_plot['회원탈퇴'])
data_plot.head() 

# (1인당 일일 중복 접속자 수는 고려하지 않음) 일일 서비스 이용 목적으로 접속한 사람의 수 (서비스 이용 접속) 
```

```python
# 데이터 타당성 
((data_plot['총가입회원']-data_plot['서비스이용접속'])<=0).sum()
```

```python
((data_plot['서비스이용접속']/data_plot['총가입회원'])*100).max()
```

```python
data_plot['서비스이용률']=(data_plot['서비스이용접속']/data_plot['총가입회원'])*100
```

### 03. ARIMA 

* 전통적인 시계열 예측 방법 
* AR(Autoregression)모델로 자기 자신의 과거를 정보로 사용
*  MA(Moving Average)모델로 이전 항에서의 오차를 이용하여 현재 항의 상태를 추론 
* 둘은 합친 것을 ARMA 👉 ARMA 모델에 추세 변동의 경향성 까지 반영한 방법이 ARIMA 
  * order = (p,d,q) 
    * p : AR 모형의 Lag 
    * d : 차분 👉 시계열 데이터의 불규칙성 보정 
    * q : MA 모형의 Lag 

#### (1) 서비스 이용률

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
series = train.서비스이용률.values
plot_acf(series)
plot_pacf(series)
plt.show() 

# 파란색 그림자 안에 bar 가 들어가있으면 통계적으로 유의하지 않다는 의미 
```

```python
import itertools 
p=q=d=range(0,4)
pdq = list(itertools.product(p,d,q)) 
params = []
aic = [] 

for param in pdq : 
  try : 
    model_arima = ARIMA(train.서비스이용률.values, order = param) 
    model_arima_fit = model_arima.fit() 
    params.append(param) 
    aic.append(model_arima_fit.aic)
  except : 
    continue 
```

```python
min(aic) # (2,1,2)
```

```python
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
model = ARIMA(train.서비스이용률.values, order = (2,1,2))
model_fit = model.fit(trend='nc', full_output=True, disp=True)
print(model_fit.summary())
 
# 상수항을 제외한 모든 계수의 p-value가 0.05 이하로 유의미한 것 -> trend = 'nc'
```

```python
fig = model_fit.plot_predict() # 학습 데이터에 대한 예측 결과이다.(첫 번째 그래프)
plt.rc('font', family='NanumGothic') 
residuals = pd.DataFrame(model_fit.resid) # 잔차의 변동을 시각화 한다. (두 번째 그래프)
residuals.plot() # 오차 변동이 매우 불안정 
```

```python
forecast_data = model_fit.forecast(steps=151)
```

```python
pred_y = forecast_data[0].tolist() 
test_y = test['서비스이용률'] 
plt.plot(pred_y, color = 'blue') 
plt.plot(test_y, color = 'red')
```
